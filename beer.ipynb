```python
def extract_name(single_listing):
    try:
        return single_listing.find_all("a")[0].text
    except:
        return None
```

```python
def extract_address(single_listing):
    try:
        return single_listing.find_all("div", class_=re.compile("address-info.*"))[0].find_all("div")[0].text
    except:
        return None
```

```python
def extract_city(single_listing):
    try:
        return single_listing.find_all("div", class_="value")[1].text
    except:
        return None
```

```python
def extract_website(single_listing):
    try:
        return single_listing.find_all("a")[2]['href']
    except:
        return None
```

```python
# build new csv file
def build_new_csv(file_name, header_names):
    file = open(file_name, 'w', newline='', encoding='utf-8')
    writer = csv.writer(file)
    headers = header_names
    writer.writerow(headers)
    file.close()

# csv_name = 'breweries.csv'
# headers = ['Name', 'Address', 'City', 'State', 'Website']
# build_new_csv(csv_name, headers)
```

```python
def update_csv(file_name, input_list):
    file = open(file_name, 'w', newline='', encoding='utf-8')
    writer = csv.writer(file)
    writer.writerow(input_list)
    file.close()
```

```python
def create_listings(url):
    # url = 'https://www.bcca.com/choose-a-state/wpbdp_category/ak/page/2'
    
    page=requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    listings = soup.find_all("div", class_ = "listing-details")
    # print(len(listing_details))
    return listings
```

```python
##   TROUBLESHOOTING 

## MAIN APP 
csv_name = 'breweries.csv'
headers = ['Name', 'Address', 'City', 'State', 'Website']
master = []
start = datetime.now()
print ("initialized at:", start.strftime("%H:%M:%S"))

for state_id in state_links.keys():
# for state_id in ['AK', 'AR']:
    print(state_id)
    
    base_url = state_links[state_id]
    
    for i in range(1,51):
        url = base_url+'page/'+str(i)
        listings = create_listings(url)
        print(url,"\n",len(listings))

        if len(listings) == 0:
            break
        else:
            for listing in listings:
                Name = extract_name(listing)
                Address = extract_address(listing)
                City = extract_city(listing)
                State = state_id
                Website = extract_website(listing)
                master.append([Name, Address, City, State, Website])
        time.sleep(1)
print("Complete!  Runtime: ", str(time.time()-datetime.now()))

# master
```

```python
url = "https://www.greatamericanbeerfestival.com/the-competition/2023-winners/"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find_all("tr")
len(results)
```

```python
df_list = []
for idx, result in enumerate(results[1:]):
    temp = result.find_all('td')
    if len(temp) == 0:
        break
    else:
        # print(idx, '-', temp[0].text)
        medal = temp[0].text
        beer_name = temp[1].text
        brewery = temp[2].text
        city = temp[3].text
        state = temp[4].text
        category = temp[5].text
        year = temp[6].text
    
        df_list.append([medal, beer_name, brewery, city, state, category, year])
```

```python
df_winners = pd.DataFrame(df_list, columns=['Medal', 'Beer Name', 'Brewery', 'City', 'State', 'Category', 'Year'])
df_winners.to_csv('winners.csv', encoding='utf-8')
```
